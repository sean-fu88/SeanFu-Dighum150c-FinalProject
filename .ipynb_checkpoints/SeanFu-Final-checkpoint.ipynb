```python
# Add all your import statements in this cell
import nltk
from nltk.corpus import brown
from nltk.tokenize.treebank import TreebankWordDetokenizer
from nltk.tokenize import sent_tokenize
import matplotlib.pyplot as plt
from collections import Counter
import requests
import sklearn
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer, TfidfVectorizer
from sklearn.decomposition import PCA, LatentDirichletAllocation
import pandas as pd
import numpy as np
```

The success of *The Moonstone* as a mystery novel, written by Wilkie Collins, is due to Collins’ use of extensive character portrayal and relationships, and the reduction of these elements leads Collins’ play adaptation *The Moonstone: A Dramatic Story, in Three Acts*, to be shallower and less effective at presenting a compelling mystery story. Due to the temporal, spatial, and physical limitations that plays have, Collins had to make significant changes to his novel to create a play adaptation. It is through the development of character details and intercharacter relationships that Collins is able keep the mystery of the missing diamond interesting, and it is through the lack of these elements that cause the play to be more bland.

Not only are the play and novel stylistically different, but there are also major plot differences. While the novel takes place over around 2 years, the play only takes place over the course of 24 hours. Additionally, the novel has its characters going to many different locations throughout the story. The play, on the other hand, happens solely at the Verinder house. Mostly due to these two limitations, the characters in the play show up at different times and in different places than in the novel with some characters not showing up at all. For example, Clack shows up to the party instead of the reader being introduced to Clack in London. The doctor Candy also is not given time to get sick and remains physically and mentally healthy throughout the play. This reduces the need for Ezra Jennings as a character and he is removed from the play. Bruff is also absent from the play. Opium is also left out of the play, with Franklin stealing the diamond while sleepwalking instead. With all these changes, Collins has greatly simplified both the plot and characters to be suitable for the play medium.

```python
narrativeBoundaries = {
    "Betteredge": [("The events related by GABRIEL BETTEREDGE", "## Second Period"),
                   ("Contributed by GABRIEL BETTEREDGE", "Epilogue")],
    "Clack": [("Contributed by MISS CLACK", "## Second Narrative")],
    "Franklin": [("Contributed by FRANKLIN BLAKE", "## Fourth Narrative"),
              ("The Story Resumed by FRANKLIN BLAKE", "## Sixth Narrative")],
    "Cuff": [("Contributed by SERGEANT CUFF", "## Seventh Narrative")],
    "Candy": [("In a Letter from MR. CANDY", "## Eighth Narrative")],
}
def getText(boundaryDict, moonstone):
    narratorText = ""
    for boundaryPair in boundaryDict:
        start, end = boundaryPair
        startLocation = moonstone.index(start)
        endLocation = moonstone.index(end)
        textChunk = moonstone[startLocation:endLocation]
        narratorText += textChunk
    return narratorText.lower()

moonstonePlay = open('Texts/MoonstonePlay.txt', encoding='utf-8').read()
#moonstone play txt from https://jhrusk.github.io/wc/moonstone.html
moonstone = moonstone = open('Texts/moonstone.md', encoding='utf-8').read()
speakerSplit = moonstonePlay.lower().split("\n") #splitting by line since each character's speech starts with "name." on a new line
narratorsSpeakers = {"Betteredge":None ,"BetteredgePlay": None, "Clack":None, "ClackPlay": None, "Franklin": None,"FranklinPlay": None, "Cuff":None,
                    "CuffPlay": None, "Candy":None, "CandyPlay": None} #combining all narrators and speakers in common from novel and play
moonstoneNarrators={"Betteredge":None, "Clack":None, "Cuff": None, "Franklin": None,"Candy":None}
playNames = {"FranklinPlay": "franklin", "CuffPlay": "cuff", "BetteredgePlay": "betteredge", "CandyPlay": "mr.",
             "ClackPlay": "clack"} #storing play names
#getting texts for moonstone novel narrators like in hw 2
for name in moonstoneNarrators:
    narratorsSpeakers[name] = getText(narrativeBoundaries[name], moonstone)
#getting texts for each character in moonstone play
for speaker, name in playNames.items():
    speech = []
    for split in speakerSplit:
        tokenizedSpeech = nltk.word_tokenize(split) #tokenizing
        if len(tokenizedSpeech)!= 0:
            if speaker == "ClackPlay": #miss sometimes refers to other characters, so looking at second token to find "clack"
                if name == tokenizedSpeech[1]:
                    for token in tokenizedSpeech:
                        speech.append(token)
            else:
                if name == tokenizedSpeech[0]: #checking if speaker's name is first token since dialogue starts with speaker
                    for token in tokenizedSpeech:
                        speech.append(token)
    narratorsSpeakers[speaker] = TreebankWordDetokenizer().detokenize(speech) #detokenizing to get a whole text for a speaker
```

In order to quantify the effects of adapting a novel into a play, a stylistic analysis can be made. Directly comparing the characters in the play to their narrator counterparts in the novel allows for a more specific inquiry into how the two different versions of *The Moonstone* differ. In the code above, both versions are split up into the texts of each narrator and the dialogue of the characters in the play, respectively, forming the corpus of study. 

```python
tfidfVectorizer = TfidfVectorizer(min_df=0.8, max_features = 400, use_idf=False)
moonstoneTextFreq = tfidfVectorizer.fit_transform(narratorsSpeakers.values()).todense()
moonstoneWords = tfidfVectorizer.get_feature_names()
moonstoneDf = pd.DataFrame(moonstoneTextFreq, index=narratorsSpeakers.keys(), columns = moonstoneWords)
pca = PCA(n_components=2)
moonstoneReduced = pca.fit_transform(moonstoneTextFreq)
moonstoneReducedDf = pd.DataFrame(moonstoneReduced, index=narratorsSpeakers.keys(), columns=['First Component', 'Second Component'])
ax = moonstoneReducedDf.plot(kind='scatter', x='First Component', y= 'Second Component', c = ['blue', 'red']*5)
for name, points in moonstoneReducedDf.iterrows():
    ax.annotate(name, points)
plt.title("Figure 1: Stylometry Analysis Between Characters")
```

Reducing the texts in each corpus to 2 PCA components based on the 400 most common words present in at least 80 percent of the corpus texts results in a stylometry analysis. By plotting this analysis, possible stylistic similarities between texts can be identified. In Figure 1, it is shown that there is a clump of texts with several other scattered points. All of the narrator texts from the novel are clustered together, signifying that they contain many stylistic similarities with each other. This makes sense as they come from one novel by one writer. On the other hand, the dialogue from the play are all vastly different from both the novel’s narrators as well as each other. The variety of styles shown in the play’s script potentially highlights how the intrinsic characteristics of plays differ from novels. Fundamentally, the difference between a play and a novel is that plays rely solely on the dialogue and acting to carry the plot and develop characters while a novel can utilize descriptive narration and character speech to engage the reader. As such, playscripts tend to be shorter and linguistically simpler than novels and such is the case for these two versions of the moonstone. Therefore, it makes sense that the play characters need to have very different styles of speaking since plays are much more primarily reliant on dialogue and acting to convey character details.  

```python
moonstoneWordTotals = {}
moonstoneAvgWord = {}
def avgNumWords(sentenceList):
    """
    Takes in a list of sentences to count only words excluding punctuation and returns the average words per sentence
    """
    numWords = 0
    numSentences = 0
    for sentence in sentenceList:
        tokenizedSentence = nltk.word_tokenize(sentence)
        numSentences += 1
        for token in tokenizedSentence:
            s = ''.join(filter(str.isalpha, token)) #replacing all non letter characters with a blank space 
            if s.lower() == '': #if token was not a word, move on to next token without counting it
                continue
            else:
                numWords += 1
    return numWords/numSentences
for name in narratorsSpeakers:
    narratorSentences = sent_tokenize(narratorsSpeakers[name])
    moonstoneAvgWord[name] = avgNumWords(narratorSentences)
    print(name, "Average Sentence Length: ", avgNumWords(narratorSentences))
def plotbar(dictname, title, xlabel, ylabel):
    bars = dictname.keys()
    y_pos = range(len(bars))
    plt.bar(y_pos, dictname.values(), color =['b', 'r'])
    # Rotation of the bars names to fit all names
    plt.xticks(y_pos, bars, rotation=90)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    return plt
plotbar(moonstoneAvgWord, "Figure 2: Average Sentence Length Per Character", "Name", "Average Sentence Length")
```

To determine how the different styles affect how the mystery of the moonstone is portrayed, a closer look at the sentences themselves is done. It can be seen that there is a large difference between the two versions. Figure 2 shows that the average words per sentence for each play speaker are significantly lower than their novel narrator counterparts. For example, in the novel when Betteredge first reunites with Franklin, Betteredge narrates in two long sentences, 

“We had had a transaction together, before he left, which consisted in his borrowing of me a ball of string, a four-bladed knife, and seven-and-sixpence in money–the colour of which last I have not seen, and never expect to see again…While he was speaking, I was looking at him, and trying to see something of the boy I remembered, in the man before me. ”

This is starkly contrasted by the play, where Franklin tells Betteredge in 4 short sentences, 

“Mr. Franklin Blake? I remember Master Franklin, the nicest boy that ever spun a top or broke a window. It’s too good to be true! …Which seven and sixpence you never have paid me back, Master Franklin, and never will.”

In the novel, Collins puts much more detail into Betteredge’s recollection of Franklin’s childhood image and debt to him when compared to the play. The small details that Betteredge recounts suggests that he cares deeply for Franklin and highlights the maturation of Franklin as an adult. While the play attempts to establish their relationship in a similar way, the more abrupt sentence structure coupled with less details makes the play relationship between Betteredge and Franklin seem more shallow in comparison. This changes the perceived dynamic between the two characters moving forward to a more servant-master relationship rather than the mentor-mentee relationship in the novel.

Not only are the sentences shorter in the play, but the average sentence length remains nearly the same between all the play speakers. Again, this is accounted for by how plays need to rely on dialogue, which is normally shorter than the long descriptive narration sentences found in novels. So, not only is there a change in the average sentence length, but there is also a change in the type of text being written about. An example of this is seen in how each version characterizes Miss Clack. In the play, Franklin speaks about Miss Clack, saying, “My old enemy?... I once called her a Rampant Spinster. Does Miss Clack still go about the world reforming everybody? And when she is particularly spiteful does she open her bag and say: ‘Permit me to offer you a tract’?” Here, Franklin directly tells the audience that Miss Clack is an intrusive, annoying person who does not like him. Meanwhile, in the novel, it is through contextual clues, speech, and actions that Collins more subtly defines Miss Clack’s character traits. The increased amount of effort required to fully understand each character not only increases the level of engagement with the reader, but it also creates more “puzzles” for the reader to solve. Now instead of just the mystery of diamond, the reader can also solve the questions: What biases does this narrator have? What information does this character know that the others don’t? What can I trust from this narrator? And unlike the play, the novel contains enough details for the reader to attempt these questions.

In the novel, the variation of average sentence length was indicative of different narrative styles as well as defining character traits. Betteredge’s narration was characteristically wordy as he was prone to distractions and rambling about various topics; Sergeant Cuff’s narration had a high average words per sentence due to his tendency to have long complex sentences explaining events and his logic which is characteristic of detective speech and writing; Mr. Candy had much shorter sentences compared to other narrators because he was mentally deteriorating from his sickness. All these nuances create more realistic characters and lead to a more captivating mystery as they diversify the story with different perspectives and styles of narration. The play loses out on these details by maintaining the same sentence length for each character. By oversimplifying the dialogue in the play, the characters do not get as much development and the story is more bland.

```python
moonstoneTagged = {}
diamondProportion= {}
narratorsSpeakersTokenized= {}
moonstoneAdjProps = {}

for name in narratorsSpeakers:
    diamondCount = 0
    narratorsSpeakersTokenized[name]= nltk.word_tokenize(narratorsSpeakers[name])
    total = len(narratorsSpeakersTokenized[name])
    moonstoneTagged[name] = nltk.pos_tag(narratorsSpeakersTokenized[name]) #adj are JJ, JJR, or JJS
    moonstoneWordTotals[name] = len(narratorsSpeakers[name])
    totalAdjective = 0
    adjList = []
    for word, kind in moonstoneTagged[name]: #for each narrator, finding the proportion and adding list of adverbs to a new dictionary
        if kind == 'JJ' or kind == 'JJR' or kind == 'JJS':
            totalAdjective +=1
    moonstoneAdjProps[name] = totalAdjective/moonstoneWordTotals[name]
    for token in narratorsSpeakersTokenized[name]:
        if token == "diamond" or token == 'moo͘nstone':
            diamondCount += 1
    diamondProportion[name] = diamondCount/total
    print(name, "Adj prop: ", moonstoneAdjProps[name], " Diamond prop: ", diamondProportion[name])
    
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plotbar(moonstoneAdjProps, 'Figure 3A: Proportion of Adjectives Per Speaker','Name','Adj Proportion')
plt.subplot(1,2,2)
plotbar(diamondProportion, "Figure 3B: Character Diamond Proportions", "Name", "Diamond Proportion")
plt.suptitle('Figure 3: Proportions for Each Character', fontsize =16)
```

The Moonstone novel is widely regarded as the first great English detective novel. Its success stemmed from Collins’ ability to keep readers engaged in the mystery of the missing diamond over a long book. One large part that contributed to this was the inclusion of different characters’ backstories, deviations not directly relating to the missing diamond, and complex character relationships (McCrum par. 7). Further analysis into the previously mentioned examples of the differences in characters between the novel and play version of the moonstone continues to show how much the play lacks in comparison to develop the characters. Figure 3 shows a comparison of different proportions seen in the novel and play. The two proportions looked at were proportion of adjectives used and proportion of using the word ‘diamond’ or ‘moonstone’. Adjective proportion was chosen because using different parts of sentence ratios can be indicative of different styles of speech. Differences in diamond proportion can also reveal shifts in the topics that characters are associated with. Figures 3A and 3B reveals that Betteredge uses less adjectives and speaks about the diamond fewer times. Looking at how this translates into his role in the play, Betteredge no longer goes on long tangents, talks about things such as *Robinson Crusoe*, or gets “detective fever” that is so characteristic of him in the novel. Rather, he is reduced to a two-dimensional servant role with his dialogue mostly consisting of calling people “sir”, “master”, and “miss” while constantly trying to be at people’s beck and call. An example that characterizes much of his dialogue in the play is, “Try again, sir, and you will find it grow on you. Shall I give you your candle?” The depth of the novel version’s Betteredge creates deeper connections between him and the other characters, resulting in more believable motives to his actions and increasing readers’ interest in the story. For example, Collins uses the moments when Betteredge catches “detective fever” as moments that the reader can empathize with and also become excited. With the shallowness of his play character, play-version Betterdge does not invoke the same emotions from the audience. Analyzing a different character– from Figures 2, 3A, and 3B– it is shown that the play-version Candy speaks just as much as other play characters, uses more adjectives than his novel counterpart, and mentions the Moonstone. This is vastly different from the novel, where Candy was suffering mentally from illness, reduced to speaking simply, and never talks about the diamond (at least in his own letter). Mr. Candy was used in the novel to include motifs such as the importance and loss of self identity and adds to the mystery with less coherent speech which must be pieced together by Ezra (Sue 133). By keeping Mr. Candy healthy and simplifying his character, the play lost out on much of the value Candy brought to the novel. 

In conclusion, from limiting the setting to simplifying the characters, Collins ends up creating an adaptation of the novel that fails to meet the standard of mystery storytelling established by his original work. Collapsing 2 years of plot into 24 hours at a single location results in character interactions that seem forced. A large factor that made *The Moonstone* engaging is the movement from one narrator’s perspective to the next. Each narrator has their own biases and limited information which creates more mystery as the reader needs to figure out what the biases are and piece together different scraps of information. This is contrasted by the play which has a much more linear and simple plot that has no main perspective to focus on at a single time. This potentially hints on why the play was not very well received and stopped production very soon after debuting. However, plays are hard to fully analyze using literary and computational analysis as plays are also reliant on the performance of the actors themselves rather than just the words in the script. Overall, the play omits much of the original plot to be more focused on the diamond, and because of that, it loses much of the nuance and character development that made the novel a captivating mystery.
